# 1 - Image Acquisition

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

    # Display the unmodified pictures
img = mpimg.imread('Side.jpg')
plt.imshow(img)
```

```python
img = mpimg.imread('Front.jpg')
plt.imshow(img)
```

# 2 - Convert to Grayscale

```python
    # Read the images using OpenCV
original_side = cv2.imread('Side.jpg')
original_front = cv2.imread('Front.jpg')

    # Average each pixel's R,G and B values
grayscale_side = (np.sum(original_side,axis=2)/3).astype(np.uint8)
grayscale_front = (np.sum(original_front,axis=2)/3).astype(np.uint8)

    # Save the grayscale images
cv2.imwrite('Side_Grey.jpg', grayscale_side)
cv2.imwrite('Front_Grey.jpg', grayscale_front)

    # Show the first image
plt.imshow(grayscale_side,cmap='gray')
```

```python
    # Show the second image
plt.imshow(grayscale_front,cmap='gray')
```

# 3 - Smooth the images using Gaussian smoothing


```python
     # Gaussian smoothing using 5x5 pixel Gaussian kernel of first image
gaussian_5_side = cv2.GaussianBlur(grayscale_side, (5,5), 0)
plt.imshow(gaussian_5_side, cmap='gray')
```

```python
    # Gaussian smoothing using 5x5 pixel Gaussian kernel of second image
gaussian_5_front = cv2.GaussianBlur(grayscale_front, (5,5), 0)
plt.imshow(gaussian_5_front, cmap='gray')
```

```python
    # Gaussian smoothing using 11x11 pixel Gaussian kernel of first image
gaussian_11_side = cv2.GaussianBlur(grayscale_side, (11,11), 0)
plt.imshow(gaussian_11_side, cmap='gray')
```

```python
    # Gaussian smoothing using 11x11 pixel Gaussian kernel of second image
gaussian_11_front = cv2.GaussianBlur(grayscale_front, (11,11), 0)
plt.imshow(gaussian_11_front, cmap='gray')
```

# 4 - Compute Image Gradients

```python
     # 5x5 pixel Gaussian kernel of first image x and y derivative
gaussian_5_side_dx = cv2.Sobel(gaussian_5_side,cv2.CV_64F,1,0,ksize=3)
gaussian_5_side_dy = cv2.Sobel(gaussian_5_side,cv2.CV_64F,0,1,ksize=3)

     # x derivative
plt.imshow(gaussian_5_side_dx, cmap='gray')
```

```python
     # y derivative
plt.imshow(gaussian_5_side_dy, cmap='gray')
```

```python
     # 11x11 pixel Gaussian kernel of first image x and y derivative
gaussian_11_side_dx = cv2.Sobel(gaussian_11_side,cv2.CV_64F,1,0,ksize=3)
gaussian_11_side_dy = cv2.Sobel(gaussian_11_side,cv2.CV_64F,0,1,ksize=3)

     # x derivative
plt.imshow(gaussian_11_side_dx, cmap='gray')
```

```python
     # y derivative
plt.imshow(gaussian_11_side_dy, cmap='gray')
```

```python
     # 5x5 pixel Gaussian kernel of second image x and y derivative
gaussian_5_front_dx = cv2.Sobel(gaussian_5_front,cv2.CV_64F,1,0,ksize=3)
gaussian_5_front_dy = cv2.Sobel(gaussian_5_front,cv2.CV_64F,0,1,ksize=3)

     # x derivative
plt.imshow(gaussian_5_front_dx, cmap='gray')
```

```python
     # y derivative
plt.imshow(gaussian_5_front_dy, cmap='gray')
```

```python
     # 11x11 pixel Gaussian kernel of second image x and y derivative
gaussian_11_front_dx = cv2.Sobel(gaussian_11_front,cv2.CV_64F,1,0,ksize=3)
gaussian_11_front_dy = cv2.Sobel(gaussian_11_front,cv2.CV_64F,0,1,ksize=3)

     # x derivative
plt.imshow(gaussian_11_front_dx, cmap='gray')
```

```python
     # y derivative
plt.imshow(gaussian_11_front_dy, cmap='gray')
```

# 5 - Compute the Edge Magnitude and Orientation

```python
    # Magnitude of 5x5 first picture
magnitude_side = np.sqrt(gaussian_5_side_dx**2 + gaussian_5_side_dy**2)
plt.imshow(magnitude_side, cmap='gray')
```

```python
    # Orientation of 5x5 first picture
orientation_side = np.arctan2(gaussian_5_side_dy, gaussian_5_side_dx)
plt.imshow(orientation_side, cmap='jet')
```

```python
    # Magnitude of 11x11 first picture
magnitude_side_11 = np.sqrt(gaussian_11_side_dx**2 + gaussian_11_side_dy**2)
plt.imshow(magnitude_side_11, cmap='gray')
```

```python
    # Orientation of 11x11 first picture
orientation_side_11 = np.arctan2(gaussian_11_side_dy, gaussian_11_side_dx)
plt.imshow(orientation_side_11, cmap='jet')
```

```python
    # Magnitude of 5x5 second picture
magnitude_front = np.sqrt(gaussian_5_front_dx**2 + gaussian_5_front_dy**2)
plt.imshow(magnitude_front, cmap='gray')
```

```python
    # Orientation of 5x5 second picture
orientation_front = np.arctan2(gaussian_5_front_dy, gaussian_5_front_dx)
plt.imshow(orientation_front, cmap='jet')
```

```python
    # Magnitude of 11x11 second picture
magnitude_front_11 = np.sqrt(gaussian_11_front_dx**2 + gaussian_11_front_dy**2)
plt.imshow(magnitude_front_11, cmap='gray')
```

```python
    # Orientation of 11x11 second picture
orientation_front_11 = np.arctan2(gaussian_11_front_dy, gaussian_11_front_dx)
plt.imshow(orientation_front_11, cmap='jet')
```

# 6 - Canny Edge Detection with opencv

```python
    # First image 5x5

canny_side = cv2.Canny(gaussian_5_side, 100, 200)
plt.figure(figsize=(10,16))
plt.imshow(canny_side, cmap='gray')
```

```python
    # First image 11x11

canny_side_11 = cv2.Canny(gaussian_11_side, 100, 200)
plt.figure(figsize=(10,16))
plt.imshow(canny_side_11, cmap='gray')
```

```python
    # Second image 5x5

canny_front = cv2.Canny(gaussian_5_front, 100, 200)
plt.figure(figsize=(10,16))
plt.imshow(canny_front, cmap='gray')
```

```python
    # Second image 11x11

canny_front_11 = cv2.Canny(gaussian_11_front, 100, 200)
plt.figure(figsize=(10,16))
plt.imshow(canny_front_11, cmap='gray')
```
